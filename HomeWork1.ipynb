```python
#Task 1
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
import seaborn as sns                           
from sklearn.neighbors import KernelDensity

n = 500000
samples_number = 1000
sample_means = []
dist_frac = 0.1

for i in range(0,samples_number):
 x1 = np.random.normal(-5,2,int(n*(1-dist_frac)))     
 x2 = np.random.normal(5,3,int(n*dist_frac))   
 x = np.concatenate((x1,x2))                       
 sample_means.append(x.mean())

plt.title("Density of {} samples".format(samples_number))
sns.kdeplot(sample_means)
sns.distplot(sample_means,color='b')
plt.legend(['seaborn sample mean','means histogram'])
```

```python
#Task 2
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import optimize

def func(x, a, b, c):
    y = np.polyval([a,b,c], x)
    return y

x = np.linspace(0, 5, 50)
noise = np.random.normal(0,0.2,50)
x = x + noise
y = func(x,np.random.random(len(x/2)),np.random.random(len(x/2)),np.random.random(len(x/2)))
alpha = optimize.curve_fit(func, xdata = x, ydata = y)[0]
print(alpha)

x_new_value = np.arange(min(x), max(x), 0.1)
y_new_value = func(x_new_value, alpha[0],alpha[1],alpha[2])

plt.figure(figsize = (10,8))
plt.plot(x,y, 'b.')
plt.plot(x_new_value,y_new_value, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
